```python
train_list=list(train['ID'])
```

```python
train_img_df=pd.DataFrame((resize(plt.imread("train/Train/"+train_list[0]),(64,64,3))/255).flatten()).T
#test_img_df=pd.DataFrame((resize(plt.imread("train/Train/"+train_list[1000]),(64,64,3))/255).flatten()).T
```

```python
for i in train_list[1:2000]:
    train_img_df=pd.concat([train_img_df,pd.DataFrame((misc.imresize(plt.imread("train/Train/"+i),(64,64,3))/255).flatten()).T],axis=0,ignore_index=True)
#for i in train_list[1001:1200]:
#    test_img_df=pd.concat([test_img_df,pd.DataFrame((misc.imresize(plt.imread("train/Train/"+i),(64,64,3))/255).flatten()).T],axis=0,ignore_index=True)
```

```python
train_img_df.shape
```

```python
def fill_target(p):
    if p=='YOUNG':
        return 1
    elif p=='MIDDLE':
        return 2
    else:
        return 3
```

```python
train['Class']=train['Class'].apply(fill_target)
```

```python
train
```

```python
train1=pd.concat([train_img_df,train['Class'].iloc[:2000]],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(train1.drop(['Class'],axis=1),train1['Class'],test_size=0.3,random_state=42)
```

# 05 : Model Selection

```python
import xgboost as xgb
from xgboost.sklearn import XGBClassifier
model=XGBClassifier()
model.fit(X_train,y_train)
```

```python
predicted=model.predict (X_test)
```

```python
predicted
```

```python
p_list=list()
for i in predicted:
    p_list.append(np.round(i))
```

```python
p_df=pd.DataFrame(p_list)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,p_df))
```
